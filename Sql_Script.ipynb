# [dbo].[sp_EmploymentVerification]

```sql
/****** Object:  StoredProcedure [dbo].[sp_EmploymentVerification]    Script Date: 6/16/2024 9:31:11 AM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE proc [dbo].[sp_EmploymentVerification]
(
@Emp_Id int,
@Company_Name nvarchar(500),
@Verification_Code nvarchar(500) 
)
as 
begin 
declare @isVerified varchar(50)=null
if exists (select 1 from EmployeeDetails where Emp_Id=@Emp_Id and Company_Name=@Company_Name and Verification_Code=@Verification_Code)
	begin 
		set @isVerified='Verified'  
	end 
else 
	begin
		set @isVerified='Not Verified'  
	end 

	select @isVerified as isVerified
end 

--exec [sp_EmploymentVerification] 10001,'Cisive','100XZ'
GO
```
